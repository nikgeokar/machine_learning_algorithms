# Correlation on GeorgeData

```python
import pandas as pd
import numpy as np
import pingouin as pg
import matplotlib.pyplot as plt
```

```python
def friedman_test(df):
    results = pg.friedman(data=df)
    q, p = results['Q'].values.tolist()[0], results['p-unc'].values.tolist()[0]
    print(f"Q={round(q,3)}, p={p}")
    # interpret
    print ('With alpha = ',0.05)
    if p > 0.001:
        # They dont have statistically significant differences between them
        print('Test Result: Same distributions\n')
    else:
        # They have statistically significant differences between them
        print('Test Result: Different distributions\n')
```

```python
geore_data = pd.read_csv('GeorgeData.csv',sep=';',index_col=0)
geore_data = geore_data.reset_index(level=0)
geore_data['Value'] = geore_data['Value']/100
corr_results = geore_data.corr()
corr_results
```

```python
friedman_test(geore_data)
```

We ran both friedman test but and also the corr function of pandas and we found that the variables are not correlated with each other.
